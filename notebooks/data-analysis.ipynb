```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
data_path = Path("../data/raw_data/raw.csv")
df = pd.read_csv(data_path)
```

```python
# First look at the dataset
print("=== FIRST OVERVIEW ===")
print(f"Number of rows: {len(df)}")
print(f"Number of columns: {len(df.columns)}")
print(f"\nColumns in the dataset:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")

print("\n=== FIRST 5 ROWS ===")
display(df.head())
```

```python
# Details about the columns
print("=== DETAILED COLUMN INFORMATION ===")
print(df.info())

print("\n=== DATA TYPES ===")
print(df.dtypes)

print("\n=== DESCRIPTIVE STATISTICS ===")
display(df.describe(include='all'))
```

```python
# Check for missing values
print("=== MISSING VALUES ===")
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_info = pd.DataFrame({
    'Column': df.columns,
    'Missing Values': missing_values,
    'Percentage (%)': missing_percentage
})

print("Missing values per column:")
display(missing_info[missing_info['Missing Values'] > 0])

if missing_info['Missing Values'].sum() == 0:
    print("No missing values found!")
else:
    print(f"Found {missing_info['Missing Values'].sum()} missing values.")
```

```python
# Check for duplicate rows
print("=== DUPLICATES ===")
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

if duplicates > 0:
    print(f"{duplicates} duplicate rows found ({duplicates/len(df)*100:.2f}%)")
    print("\nFirst few duplicates:")
    display(df[df.duplicated()].head())
else:
    print("No duplicate rows found!")

# Unique values per column
print(f"\n=== UNIQUE VALUES PER COLUMN ===")
unique_counts = pd.DataFrame({
    'Column': df.columns,
    'Unique Values': [df[col].nunique() for col in df.columns],
    'Data Type': [df[col].dtype for col in df.columns]
})
display(unique_counts)
```

```python
# Analyze numerical and categorical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

print("=== COLUMN TYPES ===")
print(f"Numerical Columns ({len(numerical_cols)}): {list(numerical_cols)}")
print(f"Categorical Columns ({len(categorical_cols)}): {list(categorical_cols)}")


print(f"\n=== NUMERICAL COLUMNS DETAILS ===")
for col in numerical_cols:
    print(f"\n{col}:")
    print(f"  - Range: {df[col].min():.2f} to {df[col].max():.2f}")
    print(f"  - Mean: {df[col].mean():.2f}")
    print(f"  - Standard Deviation: {df[col].std():.2f}")
    print(f"  - Missing Values: {df[col].isnull().sum()}")

    # Check for outliers (simple method using IQR)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)][col]
    print(f"  - Potential Outliers: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)")
```

```python
# Analyze categorical columns
print("=== CATEGORICAL COLUMNS DETAILS ===")
for col in categorical_cols:
    print(f"\n{col}:")
    print(f"  - Number of Unique Values: {df[col].nunique()}")
    print(f"  - Missing Values: {df[col].isnull().sum()}")

    # Show most frequent values
    print(f"  - Top 5 Most Frequent Values:")
    value_counts = df[col].value_counts().head()
    for value, count in value_counts.items():
        percentage = count / len(df) * 100
        print(f"    '{value}': {count} ({percentage:.1f}%)")
    
    # Warning for high cardinality
    if df[col].nunique() > len(df) * 0.8:
        print(f" High cardinality - possibly an ID column?")
```

```python
# First visualizations - distributions of numerical columns
print("=== VISUALIZATION ===")

# Calculate number of subplots
n_cols = min(3, len(numerical_cols))  # Max 3 columns
n_rows = (len(numerical_cols) + n_cols - 1) // n_cols  # Round up

plt.figure(figsize=(15, 5 * n_rows))

for i, col in enumerate(numerical_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.hist(df[col].dropna(), bins=30, alpha=0.7, edgecolor='black')
    plt.title(f'Distribution: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Correlation analysis for numerical columns
print("=== CORRELATION ANALYSIS ===")

plt.figure(figsize=(10, 8))
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap
sns.heatmap(correlation_matrix, 
            annot=True, 
            cmap='coolwarm', 
            center=0,
            fmt='.2f',
            square=True)

plt.title('Correlation matrix of numerical columns')
plt.tight_layout()
plt.show()

# Find strong correlations
print("\n Strong correlations (|r| > 0.7):")
strong_corr = []
for i in range(len(numerical_cols)):
    for j in range(i+1, len(numerical_cols)):
        corr = correlation_matrix.iloc[i, j]
        if abs(corr) > 0.7:
            strong_corr.append((numerical_cols[i], numerical_cols[j], corr))

if strong_corr:
    for col1, col2, corr in strong_corr:
        print(f"  - {col1}  {col2}: {corr:.3f}")
else:
    print("  Didn't find strong correlations.")
```
